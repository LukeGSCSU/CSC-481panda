```python
import pandas as pd
```

Create a data frame

```python
data = {
    'Name' : ['Alice', 'Bob', 'Charlie', 'David'],
    'Age' : [25, 30, 17, 28],
    'City' : ['New York', 'San Francisco', 'Los Angeles', 'Alexandria'],
    'Salary' : [100, 150, 200, 300]
}

df = pd.DataFrame(data)
df
```

```python
df[['Name','Salary']]
```

```python
df.iloc[0]
```

```python
df.iloc[1]
```

```python
df['Age'].mean()
```

```python
print("The minimum salary is", df['Salary'].min())
print("The maximum salary is", df['Age'].max())
print("The mean age is", df['Age'].mean())
```

**Series**



```python
ndata = [10, 20, 30, 40, 50]
s = pd.Series(ndata)

print(s.min())
print(s.max())
print(s.mean())
```

Apply a function on a column

```python
df['New_Salary'] = df['Salary'].apply(lambda x: x * 1.2)
df
```

Sorting the data based on a column

```python
sorted_df = df.sort_values(by = 'Age')
sorted_df
```

Create a new column based on certain conditions

```python
df['is_adult'] = df['Age'].apply(lambda x: True if x >= 18 else False)
df
```
